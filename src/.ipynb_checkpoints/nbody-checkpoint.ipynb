### Kokubo & Makino (2004) と比較

軌道要素から、初期の位置と速度を得る

```julia
a = 1.
e = 0.1
I = 0.0001
ω = Float64(π)
Ω = 0.
tₚ = Float64(π)

μ = 1.
n = √(μ / a^3)
T = 2π / n
M = - n * tₚ

elms_init = OrbitalElements(a, e, I, ω, Ω, M, tₚ, μ, n, T);
```

```julia
@show a
@show e
@show ω
@show tₚ  # denoted as T in Kokubo & Makino (2004)

@show T
@show n
@show μ;
```

```julia
t = 0
E = solveKeplerEquation2(elms_init, t)
r, v = get_rv(elms_init, E)
R = orb_to_ref(r, elms_init)
V = orb_to_ref(v, elms_init)

@show R;
@show V;
elms = OrbitalElements(R, V, elms_init.μ, t);
```

```julia
ps[2]
```

```julia
ϵ = 0.
α = 7/6
η = 0.05
Δt = 2^-5
t_end = 20π;
```

```julia
p1 = Particle(zeros(3), zeros(3), μ/G*0.5, 1.)
p2 = Particle(collect(R), collect(V), μ/G*0.5, 1.)

ps = StructArray([p1, p2])
initialize!(ps, ϵ)
```

```julia
@time times, snapshots = run_sim(ps, Δt, t_end, 1, ϵ, α, η);
```

```julia
rs = [ps[2].r - ps[1].r for ps in snapshots];
vs = [ps[2].v - ps[1].v for ps in snapshots];
```

```julia
elms_init = OrbitalElements(rs[begin], vs[begin], elms_init.μ, 0)
```

```julia
Δa = Float64[]
Δe = Float64[]
Δω = Float64[]
Δtₚ = Float64[]

for (t, r, v) in zip(times, rs, vs)
    elms = OrbitalElements(r, v, elms.μ, t)

    push!(Δa, elms.a - elms_init.a)
    push!(Δe, elms.e - elms_init.e)
    push!(Δω, elms.ω - elms_init.ω)
    push!(Δtₚ, elms.tₚ - elms_init.tₚ)
end
```

```julia
gr(bg=:white)

p1 = plot(times, Δa, legend=false, ylabel="Δa")
p2 = plot(times, Δe, legend=false, ylabel="Δe")
p3 = plot(times, Δω, legend=false, ylabel="Δω")
p4 = plot(times, Δtₚ, legend=false, ylabel="tₚ", xlabel="Time")

plot(p1, p2, p3, p4, layout=(4,1), size=(600,600), framestyle=:box,
    #xlim=(0, 10π)
)
```
